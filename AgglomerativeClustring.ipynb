```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!unzip /content/archive (2).zip
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
```

```python
df.head()
```

```python
df.drop("CustomerID", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.rename(
    columns = {
        "Spending Score (1-100)": "SpendindScore",
        "Annual Income (k$)" : "AnnualIncome"
    }, inplace = True
)
```

```python
df.head()
```

```python
df["Genre"].replace({
    "Male": 1,
    "Female": 0
},inplace=True)

df.head()
```

```python
for col in df.columns:
  sns.displot(df[col], kde = True)
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
merg = linkage(df, method='single', metric='euclidean')

plt.figure(figsize = (10, 8))
dendrogram(merg)
plt.show()
```

```python
merg = linkage(df, method='complete', metric='euclidean')

plt.figure(figsize = (10, 8))
dendrogram(merg)
plt.show()
```

```python
merg = linkage(df, method='average', metric='euclidean')

plt.figure(figsize = (10, 8))
dendrogram(merg)
plt.show()
```

```python
merg = linkage(df, method='ward', metric='euclidean')

plt.figure(figsize = (10, 8))
dendrogram(merg)
plt.show()
```

```python
from scipy.cluster.hierarchy import cut_tree

y_pred = cut_tree(merg, n_clusters=3)
```

```python
y_pred = y_pred.reshape(-1)
y_pred
```

```python
df.head(1)
```

```python
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

```python
y_pred = cut_tree(merg, n_clusters=5)
y_pred = y_pred.reshape(-1)
y_pred
```

```python
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

Can also do all the above in sklearn as below:

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='single')

y_pred = agg.fit_predict(df)
plt.figure(figsize=(10,8))
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')

y_pred = agg.fit_predict(df)
plt.figure(figsize=(10,8))
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='average')

y_pred = agg.fit_predict(df)
plt.figure(figsize=(10,8))
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

y_pred = agg.fit_predict(df)
plt.figure(figsize=(10,8))
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c = y_pred)
```

```python
import plotly.express as px

px.scatter_3d(df, x = "Age",	y = "AnnualIncome",	z = "SpendingScore", color = y_pred)
```

Lets apply this algo for circle and moon datasets

```python
from sklearn.datasets import make_circles

X , y = make_circles(n_samples=500, noise = 0.02, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
agg = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
y_pred = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_pred)
```

single linkage is working fine now bcoz of very low noise

```python
agg = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
y_pred = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_pred)
```

```python
agg = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
y_pred = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_pred)
```
