```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x , y = make_blobs(n_samples=500,centers=4,random_state=42)
print(x.shape,y.shape)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4)
km.fit(x)
```

```python
y_pred = km.predict(x)
y_pred
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y_pred)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = "red", s = 100)
```

Adding noise ......by changing standard deviation

```python
X , y = make_blobs(n_samples=500, centers=4, cluster_std=1, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
X , y = make_blobs(n_samples=500, centers=4, cluster_std=2, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
X , y = make_blobs(n_samples=500, centers=4, cluster_std=3, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=4)
km.fit(X)

y_pred = km.predict(X)

plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c = "red", s = 100)
```

**Elbow point**

WCSS VS K

inertia finds Sum of squared distances of samples to their closest cluster center, weighted by the sample weights if provided.

```python
error = []

for k in range(1, 20):
  km = KMeans(n_clusters=k)
  km.fit(X)
  error.append(km.inertia_)
```

```python
plt.plot(range(1,20),error)
```
